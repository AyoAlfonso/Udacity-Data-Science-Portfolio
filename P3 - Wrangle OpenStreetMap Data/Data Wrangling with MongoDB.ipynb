```python
# Unique tag names in the XML file 
def count_tags(filename): 
    dict = {}
    for event,elem in ET.iterparse(filename):
        tag = elem.tag
        if tag not in dict:            
            dict[tag]= 0 
        if tag in dict:
            dict[tag] = dict[tag]+1
    return dict

pprint.pprint(count_tags(OSM_FILE))
```

```python
# check the "k" value for each "<tag>" to see if they can be valid keys in MongoDB, as well as see if there are any other potential problems. 
# we have a count of each of four tag categories in a dictionary: "lower", for tags that contain only lowercase letters and are valid,
# "lower_colon", for otherwise valid tags with a colon in their names, "problemchars", for tags with problematic characters, and
# "other", for other tags that do not fall into the other three categories.

lower = re.compile(r'^([a-z]|_)*$')
lower_colon = re.compile(r'^([a-z]|_)*:([a-z]|_)*$')
problemchars = re.compile(r'[=\+/&<>;\'"\?%#$@\,\. \t\r\n]')

def key_type(element, keys):
    if element.tag == "tag":
        key_value = element.attrib["k"]
        if lower.search(key_value):
            keys["lower"] += 1
        elif lower_colon.search(key_value):
            keys["lower_colon"] += 1
        elif problemchars.search(key_value):
            keys["problemchars"] +=1
        else:
            keys["other"] +=1
    return keys

# Define a function to process the map data
def process_map(filename):
    keys = {"lower": 0, "lower_colon": 0, "problemchars": 0, "other": 0}
    for _, element in ET.iterparse(filename):
        keys = key_type(element, keys)
    return keys
```

```python
# Audit Street Type

street_type_re = re.compile(r'\b\S+\.?$', re.IGNORECASE)
expected = ["Street", "Avenue", "Boulevard", "Drive", "Court", "Place", "Square", "Lane", "Road", 
            "Trail", "Parkway", "Commons"]

def audit_street_type(street_types, street_name):
    m = street_type_re.search(street_name)
    if m:
        street_type = m.group()
        if street_type not in expected:
            street_types[street_type].add(street_name)

def is_street_name(elem):
    return (elem.attrib['k'] == "addr:street")

def audit(osmfile):
    osm_file = open(osmfile, "r")
    street_types = defaultdict(set)
    for event, elem in ET.iterparse(osm_file, events=("start",)):
        if elem.tag == "node" or elem.tag == "way":
            for tag in elem.iter("tag"):
                if is_street_name: 
                    audit_street_type(street_types, tag.attrib['v'])
    osm_file.close()
    return street_types
```

# Data Cleaning

```python
# 
# A lot of street names are ended with "Circle" or "Expressway" or "Terrace, and there is no problem with it.  
# Based on the findings, creat a mapping dictionary. 
mapping = { "St": "Street",
            "St.": "Street",
            'street':"Street",
            "Ave": "Avenue",
            'ave': "Avenue",
            "Blvd": "Boulevard",
            "Rd.": "Road",
            'Cir': "Circle",
            'Hwy':"Highway",
            'Sq':"Square",
            'Ln': "Lane",
            'Dr':"Drive",
            "W.": "West",
            "N.": "North",
            "S.": "South",
            "E": "East",
            "N": "North",
            "S": "South",
            "E": "East",
            "W": "West"}
```

```python
# Define a function to clean the street names 
def update_name(name, mapping):
    after = []
    # Split name string to test each part of the name;
    # Replacements may come anywhere in the name.
    for part in name.split(" "):
        # Check each part of the name against the keys in the correction dict
        if part in mapping.keys():
            # If exists in dict, overwrite that part of the name with the dict value for it.
            part = mapping[part]
        # Assemble each corrected piece of the name back together.
        after.append(part)
    # Return all pieces of the name as a string joined by a space.
    return " ".join(after)
```

```python
# Define a function to clean the zipcodes, only retain the elements of zipcodes in San Jose city scope
def update_postcode(zipcode):
    zipcode = zipcode.split("-")[0]
    if zipcode in sanjose_zipcode_list_str:
        return zipcode 
    elif "CA" in zipcode:
        zipcode = zipcode.split(" ")[1]
        if zipcode in sanjose_zipcode_list_str:
            return zipcode
        else:
            return False
    else:
        return False
```

#### Based on the model, in particular the following things should be done:
- process only 2 types of top level tags: "node" and "way"
- all attributes of "node" and "way" should be turned into regular key/value pairs, except:
    - attributes in the CREATED array should be added under a key "created"
    - attributes for latitude and longitude should be added to a "pos" array,
      for use in geospacial indexing. Make sure the values inside "pos" array are floats
      and not strings. 
- if the second level tag "k" value contains problematic characters, it should be ignored
- if the second level tag "k" value starts with "addr:", it should be added to a dictionary "address"
- if the second level tag "k" value does not start with "addr:", but contains ":", you can
  process it in a way that you feel is best. For example, you might split it into a two-level
  dictionary like with "addr:", or otherwise convert the ":" to create a valid key.
- if there is a second ":" that separates the type/direction of a street, the tag should be ignored.

```python
# Wrangle the data and transform the shape of the data using the code below

lower = re.compile(r'^([a-z]|_)*$')
lower_colon = re.compile(r'^([a-z]|_)*:([a-z]|_)*$')
problemchars = re.compile(r'[=\+/&<>;\'"\?%#$@\,\. \t\r\n]')

CREATED = [ "version", "changeset", "timestamp", "user", "uid"]

def shape_element(element):
    node = {}
    if element.tag == "node" or element.tag == "way" :
        nd_info = []
        address_info = {}
        node["created"] = {}
        node["id"] = element.attrib["id"]
        node["type"] = element.tag
        if "visible" in element.attrib.keys():
            node["visible"] = element.attrib["visible"]
            
        # attributes for latitude and longitude should be added to a "pos" array,
        if "lat" in element.attrib.keys():
            node["pos"] = [float(element.attrib["lat"]),float(element.attrib["lon"])]
        
        for i in CREATED:
            node["created"][i]=element.attrib[i]   
               
        for tag in element.iter("tag"):
            
            # if the second level tag "k" value contains problematic characters, it should be ignored
            if problemchars.search(tag.attrib["k"]):
                continue 
                
            # check if there is "addr:" in the tag attributes' key
            elif tag.attrib["k"].startswith("addr:"): 
                # if there is a second ":" that separates the type/direction of a street, the tag should be ignored.
                if ":" in tag.attrib["k"][5:]:
                    continue 
                else:
                    # create a key using the string after "addr:"
                    key = tag.attrib["k"][5:]
  
                    # check if the key is "street", if so, update the street name
                    if key == "street":
                        value = update_name(tag.attrib["v"], mapping)
                        address_info[key] = value                    
                    else:
                        address_info[key] = tag.attrib["v"]
                    
                    if address_info:
                        node["address"] = address_info
            else:
                node[tag.attrib['k']] = tag.attrib['v']
        
        # for "way" specifically, turn the value of nd 'ref' key into a list   
        for tag2 in element.iter("nd"):
            nd_info.append(tag2.attrib['ref'])
        
        if nd_info != []:
            node['node_refs'] = nd_info
        
        # check if there is "postcode" in the element, if so, check whether it is within the scope of San Jose. If so, 
        # standardize the postcode. Otherwise, ignore this element. 
        if "address" in node:
            if "city" in node["address"]:
                if node["address"]["city"] not in ["San Jose", "san jose",'san Jose','San jose']:
                    return None 
            if "postcode" in node["address"]:
                value = node["address"]["postcode"]
                if update_postcode(value) is False:
                    return None  
                else:
                    value = update_postcode(value)
                    node["address"]["postcode"] = value
        return node
```

```python
# turn into json 
def process_file(file_in, pretty = False):
    
    file_out = "{0}.json".format(file_in)
    data = []
    with codecs.open(file_out, "w") as fo:
        for _, element in ET.iterparse(file_in):
            el = shape_element(element)
            if el:
                data.append(el)
                if pretty:
                    fo.write(json.dumps(el, indent=2)+"\n")
                else:
                    fo.write(json.dumps(el) + "\n")
    return data
```

# Benefits and problems with additional improvements discussed

### 1.Enhance format standardization:
**Benefits:** 
- saving huge amount of time in terms of data munging since information is more aligned in format; 
- for users who creating map data, it's also easier for them to edit the info since there is a gold standard to follow and auto-correction exist to aid the process

**Problems:**
- may not everyone could get used to the enforced format, it's hard to find a gold standard for all kinds of info and for every one to follow due to regional habits & preference. 
- Also, since OpenStreetMap is a global platform, it means a lot of workload to get it done considering the massive number of countries & languages. 

### 2.Autofilter for mistaken information:
**Benefits:** 
- avoiding some obvious mistake to enhance the data quality.

**Problems:**
- need to preset the geoinfo association with zipcode and city name, and constantly update the changes. This requires a lot of workload, and may not applicable for every country. 

### 3.User contribution:
**Benefits:** 
- more users involved will help to develop a better community, thus the map data will be not only more complete but also more useful & powerful.   

**Problems:**
- current method of collecting and inputing data is not that easy so everyone could execute. Need time and energy to develop better interface and method for users to follow. 
- Also, it's quite capital intensive to extend the functions for a non-profit website which is currently rely on volunteers. There will be a huge challenge to fund this initiative.
